```python
import pandas as pd
import yfinance as yf
from datetime import datetime
from tqdm import tqdm
```

```python
start_date = datetime(2021, 1, 1)
end_date = datetime(2024, 11, 15)
```

```python
stock_list = [
   "BACQU", "NTWOU", "GSRTU", "ADUR", "WLACU", "ASPCU", "ACOG", "IZTC", "FBLA", "WRD"
]
```

```python
percentage_changes = pd.DataFrame()

for equity_name in tqdm(stock_list, desc="Downloading data"):
    equity_data = yf.download(equity_name, start=start_date, end=end_date)
    pct_change = equity_data['Adj Close'].pct_change().dropna()
    print(pct_change[0:1])
    percentage_changes[equity_name] = pct_change
```

```python
correlation_matrix = percentage_changes.corr()
display(correlation_matrix)
```
